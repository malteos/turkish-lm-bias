# Figures for LREC paper

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("./p_values.csv", index_col=0)
df
```

```python
df = df.drop(columns=["selected"])
```

```python
"""
insignifacnt (p > 0.1)
margianally significant  (p < 0.1)
significant (p < 0.05)
highly significant (p < 0.01)
"""
p_value_internals = {
    "insignificant": {
        "min_p": 0.1,
        "color": "",
    },
    "marginally_significant": {
        "p_max": 0.1,
        "color": "",
    },
    "significant": {
        "p_max": 0.05,
        "color": "",
    },
    "highly_significant": {
        "p_max": 0.01,
        "color": "",
    }
}
```

```python
import matplotlib.colors

from matplotlib.colors import BoundaryNorm, ListedColormap
from matplotlib.lines import Line2D
```

```python
plt.tight_layout()

#unavail_color = 'lightgray'

my_labels = [
    'highly significant (p < 0.01)', 
    'significant (p < 0.05)', 
    'marginally sig. (p < 0.1)', 
    'not significant (p > 0.1)',  
] # 'unavailable'
my_colors = ['darkblue', 'dodgerblue', 'lightskyblue', 'whitesmoke']  # , dodgerblluee, lightskyblue
my_cmap = ListedColormap(my_colors)
bounds = [0, 0.01, 0.05, 0.10, 1.0]
my_norm = BoundaryNorm(bounds, ncolors=len(my_colors))

#mycmap = matplotlib.colors.ListedColormap(['yellow','orange','red','darkred'])

selected_idxs = {
    'weat6-turkish': "WEAT 6", 
    'weat6b-turkish': "WEAT 6b", 
    'weat6-turkish_religious': "WEAT 6 (religious)",
    'weat8-turkish': "WEAT 8", 
    'weat8b-turkish': "WEAT 8b",
    #
    'seat4': 'SEAT 4',
    'sent-weat6-turkish': "SEAT 6", 
    'sent-weat6b-turkish': "SEAT 6b", 
    'sent-weat8-turkish': "SEAT 8", 
    'sent-weat8b-turkish': "SEAT 8b",

    "heilman_double_bind_competent_1+3-turkish": "DB competent 1+3",
    "heilman_double_bind_competent_1+3-turkish_religous": "DB competent 1+3 (religious)",
    "heilman_double_bind_competent_1-turkish": "DB competent 1",
    "heilman_double_bind_competent_1-turkish_religous": "DB competent 1  (religious)",
    "sent-heilman_double_bind_competent_one_word-turkish": "DB competent one word",
    "sent-heilman_double_bind_competent_one_word-turkish_religous": "DB competent one word  (religious)",
    "sent-heilman_double_bind_likable_one_word-turkish": "DB likeable one word",
    "sent-heilman_double_bind_likable_one_word-turkish_religous": "DB likeable one word (religious)",
}

filtered_df = df[df.index.isin(set(selected_idxs.keys()))].T

fig, ax = plt.subplots(figsize=(10,3))  # (10,18)

sns.heatmap(filtered_df, 
            annot=True, 
            linewidth=.5, 
            cmap=my_cmap,
            norm=my_norm,
            square=False, 
            ax=ax,  fmt='.2f', cbar_kws={"shrink": 0.5}, cbar=False)
ax.set(xlabel="", ylabel="")
#ax.tick_params(axis='x', rotation=45)

model_names = [
    "BERTurk-cased 32k",
    "BERTurk-uncased 32k",
    "BERTurk-cased 128k",
    "BERTurk-uncased 128k",
    "mBERT-cased",
    "mBERT-uncased",
    "mT5-base",
]

ylabels = model_names
xlabels = selected_idxs.values()

#xlabels = model_names
#ylabels = selected_idxs.values()

ax.set_yticklabels(
    labels=ylabels,
)

ax.set_xticklabels(
    labels=xlabels, #df.columns, 
    rotation=45, 
    #ha='left', 
    ha='right', 
    rotation_mode='anchor') ### based offset
#ax.xaxis.tick_top()


handles = [Line2D([], [], lw=10, color=color, label=label)
           for color, label in zip(my_colors, my_labels)] #  + [unavail_color]
ax.legend(handles=handles, 
          handlelength=0.5, 
          ncol=len(my_colors)/2,  #  + 1
          #bbox_to_anchor=(0.5, -0.02), 
          bbox_to_anchor=(0.5, 1.3),
          loc='upper center', 
          frameon=False
         )

plt.savefig('figure1_groupterms.pdf', bbox_inches='tight')
```
